```python
import pandas as pd
import numpy as np
import dataframe_image as dfi
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import csv
```

```python
df = pd.read_csv("ShowcaseDataWinter2023.csv").dropna()
```

```python
df.dtypes
```

```python
for val in ['online_appraisal_flag', 'model_year', 'mpg_city', 'mpg_highway', 'fuel_capacity', 
            'model_year_appraisal', 'mpg_city_appraisal', 'fuel_capacity_appraisal']:
    df[val] = df[val].astype('int64')

cleanup_trim = {"trim_descrip": {"Premium": 1, "Not Premium": 0},
               "trim_descrip_appraisal": {"Premium": 1, "Not Premium": 0}}

df = df.replace(cleanup_trim)

df['engine'] = df['engine'].str.slice_replace(start = -1, repl = '').astype('float64')
df['engine_appraisal'] = df['engine_appraisal'].str.slice_replace(start = -1, repl = '').astype('float64')
```

```python
def rangeSplit(df, columns):
    new_cols = []
    for col in columns:
        lo = "lo_" + col
        hi = "hi_" + col
        cleanedString = df[col].str.replace('k', '', regex = True).str.replace('\$', '', regex = True).str.replace('+', '', regex = True).str.replace(' miles', '', regex = True)
        splitRange = cleanedString.str.split(r" to ")
        df[lo] = splitRange.str[0].astype('Int64')
        df[hi] = splitRange.str[1].astype('Int64')
    return df
```

```python
rangeCols = ['price', 'appraisal_offer', 'mileage', 'mileage_appraisal']
df = rangeSplit(df, rangeCols)
```

```python
feature_cols = ['model_year_appraisal', 'trim_descrip_appraisal', 'body_appraisal', 'engine_appraisal',
                'cylinders_appraisal','mpg_city_appraisal', 'mpg_highway_appraisal', 'horsepower_appraisal',
                'fuel_capacity', 'lo_appraisal_offer', 'hi_appraisal_offer', 'lo_mileage_appraisal',
                'hi_mileage_appraisal']
```

```python
feat_df = df[feature_cols]
```
